```python
from google.colab.patches import cv2_imshow    # This is only is used for to show the image.
```

```python
import cv2
```

```python
image = cv2.imread('/rashmika.jpg')
```

```python
cv2_imshow(image)
```

```python
image_gray = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(image_gray)
```

```python
invertedblur = cv2.bitwise_not(image_gray)
```

```python
cv2_imshow(invertedblur)
```

```python
blur_image = cv2.GaussianBlur(invertedblur, (21,21) , sigmaY = 0, sigmaX = 0)
cv2_imshow(blur_image)
```

```python
def dodgeV2(x,y):
  return cv2.divide(x ,255-y ,scale = 256)
  # Dodging and Burning is a old photographic techniques.
```

```python
final_image = dodgeV2(image_gray , blur_image)
cv2_imshow(final_image)
```
